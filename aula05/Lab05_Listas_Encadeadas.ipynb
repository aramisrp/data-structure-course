## 1. A Classe `Node`
O `Node` (Nó) é o bloco de construção básico da nossa lista. Diferente de um vetor, ele deve obrigatoriamente guardar duas informações:
1. O **Valor** (dado).
2. O **Ponteiro** (`next`), que guarda o caminho para o próximo nó. Ao ser instanciado, ele nasce "cego", apontando para `None`.

```python
class Node:
    """Classe que representa um único nó (elemento) da lista encadeada."""
    
    def __init__(self, valor):
        self.valor = valor
        self.next = None  # O nó nasce "solto" na memória, sem apontar para ninguém
```

## 2. A Classe `LinkedList`
A lista encadeada em si não guarda todos os dados, ela precisa conhecer apenas a "cabeça" da estrutura (`head`). Para inserir no final ou imprimir, o processador precisa caminhar nó por nó utilizando um laço de repetição (`while`).

```python
class LinkedList:
    """Classe que gerencia os nós, formando a lista encadeada simples."""
    
    def __init__(self):
        self.head = None  # A lista começa vazia
        
    def append(self, valor):
        """Insere um novo nó no final da lista."""
        novo_no = Node(valor)
        
        # Se a lista estiver vazia, a cabeça passa a ser o novo nó
        if self.head is None:
            self.head = novo_no
            return
            
        # O Motor de Busca: caminhar até o final da lista
        current = self.head
        while current.next is not None:
            current = current.next
            
        # Regra de Ouro: amarramos o último nó ao novo nó que acabou de chegar
        current.next = novo_no
        
    def display(self):
        """Percorre a lista imprimindo os valores para comprovar o encadeamento."""
        elementos = []
        current = self.head
        
        # Caminha pela lista coletando os dados
        while current is not None:
            elementos.append(str(current.valor))
            current = current.next
            
        elementos.append("None")
        print(" -> ".join(elementos))
```

## 3. Teste de Mesa
Chegou a hora de instanciar a nossa classe, inserir 3 valores e imprimir a estrutura no terminal para verificar se os ponteiros foram costurados corretamente.

```python
# Instanciando a lista na memória
minha_lista = LinkedList()

# Inserindo 3 valores aleatórios
minha_lista.append(10)
minha_lista.append(20)
minha_lista.append(30)

# Verificando a saída visual
print("Estrutura atual da memória:")
minha_lista.display() 
# A saída esperada deve ser: 10 -> 20 -> 30 -> None
```
